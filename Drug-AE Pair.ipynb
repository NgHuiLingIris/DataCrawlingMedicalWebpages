```python
import pandas as pd

data = pd.read_csv('Drug-AE.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
a_df = data[['INGREDIENT','MEDDRA_PREFERRED_TERM','REFERENCE_NO']].groupby(['INGREDIENT','MEDDRA_PREFERRED_TERM'], as_index = False).count()
a_df.head()
```

```python
aplusb_df = data[['INGREDIENT','REFERENCE_NO']].groupby(['INGREDIENT'],as_index=False).count()
aplusb_df.head()
```

```python
aplusc_df = data[['MEDDRA_PREFERRED_TERM','REFERENCE_NO']].groupby(['MEDDRA_PREFERRED_TERM'],as_index=False).count()
aplusc_df.head()
```

```python
total = data['REFERENCE_NO'].drop_duplicates().count()
total
```

```python
ROR_df = data[['INGREDIENT','MEDDRA_PREFERRED_TERM']].drop_duplicates()
ROR_df
```

```python
overall_df = pd.merge(ROR_df, a_df,how='left', left_on=['INGREDIENT','MEDDRA_PREFERRED_TERM'], right_on=['INGREDIENT','MEDDRA_PREFERRED_TERM'])
overall_df = overall_df.rename(columns={'REFERENCE_NO' : 'a'})
overall_df.head()
```

```python
overall_df = overall_df.merge(aplusb_df,how='left', left_on=['INGREDIENT'], right_on=['INGREDIENT'])
overall_df = overall_df.rename(columns={'REFERENCE_NO' : 'a_plus_b'})
overall_df['b'] = overall_df['a_plus_b'] - overall_df['a']
overall_df.head()
```

```python
overall_df = overall_df.merge(aplusc_df,how='left', left_on=['MEDDRA_PREFERRED_TERM'], right_on=['MEDDRA_PREFERRED_TERM'])
overall_df = overall_df.rename(columns={'REFERENCE_NO' : 'a_plus_c'})
overall_df['c'] = overall_df['a_plus_c'] - overall_df['a']
overall_df.head()
```

```python
overall_df['d'] = total - overall_df['a'] - overall_df['b'] - overall_df['c']
overall_df.head()
```

```python
overall_df['ROR'] = (overall_df['a'] + 0.5) * (overall_df['d'] + 0.5) / (overall_df['b'] + 0.5) * (overall_df['c'] + 0.5)
overall_df
```

```python
import numpy as np
overall_df['SD_ROR'] = np.sqrt(1/(overall_df['a'] + 0.5) + 1/(overall_df['d'] + 0.5) + 1/(overall_df['b'] + 0.5) + 1/(overall_df['c'] + 0.5))
overall_df
```

```python
overall_df['ROR â€“ 1.96*SD'] = overall_df['ROR']-(1.96*overall_df['SD_ROR'])
overall_df
```

```python
flag1 = overall_df[overall_df['a']>3]
flag2 = overall_df[overall_df['ROR â€“ 1.96*SD']>1]

flagged_df= pd.concat([flag1,flag2]).drop_duplicates()

flagged_df
```
